```python
import re

filename = "D:\\documents\\jupyter\\vis final\\caption.ass"
with open(filename, encoding="utf-8") as f:
    content = f.readlines()
# you may also want to remove whitespace characters like `\n` at the end of each line
content = [x.strip() for x in content]
```

```python
timeResult = []
for line in content:
    if line.find("Dxw68") != -1:
        timeResult.append(
            {
                "time": line[12:22],
                "speech": re.sub(r"-", "", ((line[line.find("E 32}") + 5 :]))).strip(),
            }
        )
```

```python
import json
scenes=json.load(open('scenes.json'))

timeResultPtr=0

lefts=[]

for scene in scenes:
    for line in scene['sceneSpeech']:
        lefts.append(line['speech'])
        
rights=[i['speech'] for i in timeResult]


```

```python
leftPtr=0
rightPtr=0
pair=[]
def ins(l,r):
    pair.append({
            'left':l,
            'right':r
        })
while True:
    for err in range(-1,20):
        if err>15:
            print('ERROR!')
            print('left: \t',lefts[leftPtr])
            print('right: \t',rights[rightPtr])
        if rightPtr+err>=0 and sim(lefts[leftPtr],rights[rightPtr+err])>0.5:
            rightPtr+=err
            ins(lefts[leftPtr],rights[rightPtr])
            print('err=',err,'sim=', sim(lefts[leftPtr],rights[rightPtr]))
            break
        if leftPtr+err>=0 and sim(lefts[leftPtr+err],rights[rightPtr])>0.5:
            leftPtr+=err
            ins(lefts[leftPtr],rights[rightPtr])
            print('err=',err,'sim=', sim(lefts[leftPtr],rights[rightPtr]))
            break
    print('left: \t',lefts[leftPtr])
    print('right: \t',rights[rightPtr])
    print()
    leftPtr+=1
    rightPtr+=1
```

```python
import re, string
pList=str.maketrans({i:'' for i in string.punctuation})
def sim(x,y):
    xBag=x.translate(pList).lower().split()
    yBag=y.translate(pList).lower().split()
#     print(xBag,yBag)
    score=0
    for xWord in xBag:
        if xWord in yBag:
            score+=1
    scoreX=score/len(xBag)
    score=0
    for yWord in yBag:
        if yWord in xBag:
            score+=1
    scoreY=score/len(yBag)
    return max(scoreX,scoreY)
sim("It's all right. It's all right."
,"It's just,")
```
